```python
# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ –∏ –Ω–∞—Å—Ç—Ä–æ–π–∫–∞
!pip install ollama chromadb sentence-transformers

# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ Ollama
!curl -fsSL https://ollama.com/install.sh | sh

# –ó–∞–ø—É—Å–∫ Ollama –≤ —Ñ–æ–Ω–µ
import subprocess
import threading
import time

def run_ollama():
    subprocess.run(["ollama", "serve"])

# –ó–∞–ø—É—Å–∫–∞–µ–º –≤ –æ—Ç–¥–µ–ª—å–Ω–æ–º –ø–æ—Ç–æ–∫–µ
ollama_thread = threading.Thread(target=run_ollama)
ollama_thread.daemon = True
ollama_thread.start()

# –ñ–¥–µ–º –∑–∞–ø—É—Å–∫–∞
time.sleep(5)

# –ü–æ–¥–∫–ª—é—á–∞–µ–º Google Drive
from google.colab import drive
drive.mount('/content/drive')

# –ü—É—Ç—å –∫ –≤–∞—à–∏–º –¥–∞–Ω–Ω—ã–º
DATA_PATH = "/content/drive/MyDrive/json_calls"  # –ò–∑–º–µ–Ω–∏—Ç–µ –Ω–∞ —Å–≤–æ–π –ø—É—Ç—å

# –ü—Ä–æ–≤–µ—Ä—è–µ–º –¥–∞–Ω–Ω—ã–µ
import os
if os.path.exists(DATA_PATH):
    print(f"‚úÖ –î–∞–Ω–Ω—ã–µ –Ω–∞–π–¥–µ–Ω—ã: {len(os.listdir(DATA_PATH))} —Ñ–∞–π–ª–æ–≤")
else:
    print(f"‚ùå –î–∏—Ä–µ–∫—Ç–æ—Ä–∏—è –Ω–µ –Ω–∞–π–¥–µ–Ω–∞: {DATA_PATH}")
    # –°–æ–∑–¥–∞–µ–º —Ç–µ—Å—Ç–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ –µ—Å–ª–∏ –Ω–µ—Ç
    os.makedirs(DATA_PATH, exist_ok=True)
    print("üìù –°–æ–∑–¥–∞–Ω–∞ —Ç–µ—Å—Ç–æ–≤–∞—è –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—è")

# –°–∫–∞—á–∏–≤–∞–µ–º –ª–µ–≥–∫—É—é –º–æ–¥–µ–ª—å
!ollama pull qwen2.5:3b  # –ò–ª–∏ phi3:mini, mistral:7b

# –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –Ω–∞—à—É —Å–∏—Å—Ç–µ–º—É
sys.path.append('/content')

from mcp_orchestrator import ColabJSONCallAnalyticsMCP

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å–∏—Å—Ç–µ–º—É
print("üöÄ –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è —Å–∏—Å—Ç–µ–º—ã...")
system = ColabJSONCallAnalyticsMCP(DATA_PATH, "qwen2.5:3b")

# –¢–µ—Å—Ç–æ–≤—ã–π –∑–∞–ø—Ä–æ—Å
print("\nüß™ –¢–µ—Å—Ç–æ–≤—ã–π –∑–∞–ø—Ä–æ—Å...")
result = system.process_query("–°–∫–æ–ª—å–∫–æ –∂–∞–ª–æ–± –Ω–∞ –∫–∞—á–µ—Å—Ç–≤–æ –≤ —ç—Ç–æ–º –º–µ—Å—è—Ü–µ?")
print(f"–û—Ç–≤–µ—Ç: {result['answer']}")
print(f"–í—Ä–µ–º—è: {result['processing_time']}")

# –ò–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ã–π —Ä–µ–∂–∏–º
def interactive_mode():
    print("\n" + "="*50)
    print("ü§ñ MCP –ê–Ω–∞–ª–∏—Ç–∏–∫–∞ –∑–≤–æ–Ω–∫–æ–≤ - Colab –≤–µ—Ä—Å–∏—è")
    print("="*50)
    
    while True:
        query = input("\nüìù –í–∞—à –∑–∞–ø—Ä–æ—Å (–≤—ã—Ö–æ–¥ –¥–ª—è –∑–∞–≤–µ—Ä—à–µ–Ω–∏—è): ").strip()
        
        if query.lower() in ['–≤—ã—Ö–æ–¥', 'exit', 'quit']:
            print("üëã –î–æ —Å–≤–∏–¥–∞–Ω–∏—è!")
            break
        
        if not query:
            continue
            
        print("üîç –û–±—Ä–∞–±–∞—Ç—ã–≤–∞—é...")
        result = system.process_query(query)
        print(f"\nüí° –û—Ç–≤–µ—Ç: {result['answer']}")
        print(f"‚è±Ô∏è  –í—Ä–µ–º—è –æ–±—Ä–∞–±–æ—Ç–∫–∏: {result['processing_time']}")

# –ó–∞–ø—É—Å–∫–∞–µ–º –∏–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ã–π —Ä–µ–∂–∏–º
interactive_mode()
```
